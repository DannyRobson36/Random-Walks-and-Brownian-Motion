```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(n):
    steps = np.random.choice([-1, 1], size=n, p=[0.5, 0.5])
    a = np.cumsum(steps)
    b = np.insert(a,0,0)
    return b

# Set the number of steps
N = 100

# Generate random walk data
random_walk_1 = random_walk(N)
random_walk_2 = random_walk(N)
random_walk_3 = random_walk(N)
# Plot the random walk
plt.plot(range(0, N+1), random_walk_1, color = 'black')
plt.plot(range(0, N+1), random_walk_2, color = 'black')
plt.plot(range(0, N+1), random_walk_3, color = 'black')
plt.xlabel('Number of Steps (N) (p = 0.5)')
plt.ylabel('Value of S_n')
plt.grid(True)
plt.show()
```
