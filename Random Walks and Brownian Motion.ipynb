### Random Walks and Brownian Motion

Here is the code used to produce 'Figure 1' and 'Figure 2' in my essay 'Random Walks and Brownian Motion. \
Note that due to the random nature of this code, the graphs below will look different to the screenshots used in the essay, as each time the code is run a new graph is produced.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (9,6))

def random_walk_symmetric(n): # Random Walk function
    steps = np.random.choice([-1, 1], size=n, p=[0.5, 0.5]) # Creates an array which contains a random choice between +1 and -1 in each entry,
                                                            # with equal probabilities for each (0.5 chance of -1, 0.5 chance of 1)
    a = np.cumsum(steps) # Turns the array into a cumulative sum of its elements - captures the path of the random walk
    b = np.insert(a,0,0) # Inserts 0 as the first entry in the array
    return b

N = 100 # Set the number of steps in the random walk

random_walk_1 = random_walk_symmetric(N) # Generate 3 instances of the random walk
random_walk_2 = random_walk_symmetric(N)
random_walk_3 = random_walk_symmetric(N)

plt.plot(range(0, N+1), random_walk_1, color = 'black') # Plot the 3 random walks side by side
plt.plot(range(0, N+1), random_walk_2, color = 'black')
plt.plot(range(0, N+1), random_walk_3, color = 'black')
plt.xlabel('Number of Steps (N) (p = 0.5)')
plt.ylabel('Value of S_n')
plt.grid(True)
plt.show()
```
